The next sections will help you through the process of setting the parameters up to connect and retrieve the data.

## Settings

In order to fetch data from the data retrieval instance, you'll need to set some parameters regarding its host machine. 

### Configuring the IP address and port
First, define its **IP address** and the **TCP port** reserved for the servlet in the textbox below, following the format `address:port`. Press [Enter] or push the update button to submit your change.

Also note that, by default, it is already set to access the Controls Group's server instance (`10.0.6.57:8080`).

```python
epicsArchiverWidgets.drawAddressWidgets ()
```

### Configuring the connection type

Next, we need define if the connection will encrypted with the HTTPS application protocol. By default, as we are going to use a self-signed SSL certificate, this option is checked.

```python
epicsArchiverWidgets.drawConnectionWidgets ()
```

## Fetching the data

Once the settings were finished, we can now fetch data from the archiver. Select **start** and **end** dates:

```python
epicsArchiverWidgets.drawDatetimeWidgets ()
```

You need to specify the PV names and if the query should optimized or not:

```python
epicsArchiverWidgets.drawVariableWidgets()
```

### Plotting

There are some functions that can help you to plot your archived data. The following command plots all listed variables into a single figure:

```python
epicsArchiverWidgets.drawPlot (variables = epicsArchiverWidgets.getVariables())
```

If you want to plot only the `i-th` variable, use the command below:

```python
i = 0
epicsArchiverWidgets.drawPlot (variables = epicsArchiverWidgets.getIthVariable (i))
```

Finally, if you need to plot a part of the variables, use a list of the variable indexes:

```python
pvlist = [0,1]
epicsArchiverWidgets.drawPlot (variables = epicsArchiverWidgets.getRangeVariables(pvlist))
```

If you need more than a figure, use the `group` param to group your variables into plots.

```python
group1 = epicsArchiverWidgets.getRangeVariables([0, 1])
group2 = epicsArchiverWidgets.getRangeVariables([2, 3])
epicsArchiverWidgets.drawPlot (groups = [ group1 group2 ])
```
